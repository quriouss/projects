```python
import tensorflow as tf
from glob import glob
import pandas as pd
import cv2
from builtins import range, input
from tensorflow import keras
from tensorflow.keras.layers import Input, Lambda, Dense, Flatten, GlobalAveragePooling2D, Dropout
from tensorflow.keras.models import Model, load_model
from tensorflow.keras.applications import DenseNet201
from tensorflow.keras.applications.densenet import preprocess_input
from tensorflow.keras.preprocessing import image
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from keras.models import Model,Sequential,load_model
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelBinarizer
from tensorflow.keras.utils import to_categorical
from sklearn.metrics import confusion_matrix, roc_curve
```

```python
# Define paths
covid_path = ".\\kaggle\\COVID"
noncovid_path = ".\\kaggle\\non-COVID"

# Use glob to retrieve all pathnames 
covid_files = glob(covid_path + '\*')
noncovid_files = glob(noncovid_path + '\*')
```

## Προεπεξεργασία Δεδομένων (Data Preprocessing)
Οι εικόνες παρουσίαζουν μεγάλο μέγεθος γεγονός που καθιστά αδύνατη τη φόρτωση τους στο νευρωνικό δίκτυο. Για να μπορέσουν να δοθούν ως είσοδος στο νευρωνικό δίκτυο μετασχηματίστηκαν και έλαβαν το μέγεθος 224x224. Επιπλέον, τα εικονοστοιχεία (pixels) των εικόνων παρουσιάζουν εύρος τιμών στο διάστημα [0, 255]. Κάθε εικονοστοιχείο της εικόνας διαιρέθηκε με την τιμή 255 ώστε να λαμβάνει τιμές στο διάστημα [0, 1]. Σε δεύτερο στάδιο, οι covid και οι non-covid εικόνες χωρίστηκαν σε σύνολο εκπαίδευσης (train set) και σε σύνολο δοκιμής (test set) με αναλογία 80:20 του αρχικού συνόλου δεδομένων, αντίστοιχα. Στη συνέχεια, το σύνολο εκπαίδευσης των covid εικόνων και των non-covid εικόνων ενώθηκαν για να σχηματίσουν ένα κοινό σύνολο εκπαίδευσης. Η ίδια διαδικασία ακολουθήθηκε και για το σύνολο ελέγχου. 

```python
#define size to which images are to be resized
IMAGE_SIZE = [224, 224]

# Fetch Images and Class Labels from Files
covid_labels = []
noncovid_labels = []

covid_images=[]
noncovid_images=[]

for i in range(len(covid_files)):
  image = cv2.imread(covid_files[i]) # read file 
  image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB) # arrange format as per kera
  image = cv2.resize(image, IMAGE_SIZE) # resize as per model
  covid_images.append(image) # append image
  covid_labels.append('CT_COVID') # append class label

# The same procedure for noncovid files
for i in range(len(noncovid_files)):
  image = cv2.imread(noncovid_files[i])
  image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
  image = cv2.resize(image, IMAGE_SIZE)
  noncovid_images.append(image)
  noncovid_labels.append('CT_NonCOVID')
```

```python
# Convert to array and normalize to interval of [0,1]
covid_images = np.array(covid_images) / 255
noncovid_images = np.array(noncovid_images) / 255
```

```python
# Split into train and test sets for both types of images
covid_x_train, covid_x_test, covid_y_train, covid_y_test = train_test_split(
    covid_images, ['CT_Covid' for file in covid_files], test_size=0.2)

noncovid_x_train, noncovid_x_test, noncovid_y_train, noncovid_y_test = train_test_split(
    noncovid_images, ['CT_NonCovid' for file in noncovid_files], test_size=0.2)

# Merge sets for both types of images
X_train = np.concatenate((noncovid_x_train, covid_x_train), axis=0)
X_test = np.concatenate((noncovid_x_test, covid_x_test), axis=0)

y_train = np.concatenate((noncovid_y_train, covid_y_train), axis=0)
y_test = np.concatenate((noncovid_y_test, covid_y_test), axis=0)

# Make labels into categories - either 0 or 1
y_train = LabelBinarizer().fit_transform(y_train)
y_train = to_categorical(y_train)

y_test = LabelBinarizer().fit_transform(y_test)
y_test = to_categorical(y_test)
```

## Κατασκευή του Νευρωνικού Δικτύου 
Πριν  από  την εκπαίδευση του δικτύου, ήταν απαραίτητο να κατασκευαστεί και να περιγραφεί η αρχιτεκτονική του. Το προεκπαιδευμένο δίκτυο φορτώθηκε μαζί με τα βάρη του τα οποία χρησιμοποιήθηκαν αυτούσια. Έπειτα, αφαιρέθηκε το τελευταίο πλήρως συνδεδεμένο στρώμα που δίνει την έξοδο του νευρωνικού, δηλαδή την πρόγνωση. Αυτό επιτεύχθηκε με την ανάθεση της μεταβλητής  «include_top» στην τιμή False. Ακολούθησε η προσθήκη πέντε νέων επιπέδων ώστε να εκπαιδευτούν με βάση το παρόν σύνολο δεδομένων. Πιο συγκεκριμένα, εφαρμόστηκε η "Καθολική Υποδειγματοληψία Μέσου όρου (Global Average Pooling)"  η οποία μειώνει το μέγεθος των δεδομένων, το πλήθος των παραμέτρων, συμβάλοντας στην ελάττωση του υπολογιστικού κόστους. Στη συνέχεια, ακολούθησε ένα «επίπεδο ισοπέδωσης» (flattening layer) το οποίο αναδομεί τα δεδομένα ξεδιπλώνοντας τους χάρτες γνωρισμάτων εξόδου ενός συνελικτικού δικτύου και μετατρέποντας τα σε ένα μονοδιάστατο πίνακα. Στο τρίτο επίπεδο, προστέθηκε ένα πλήρως συνεδεμένο στρώμα με 128 νευρώνες και με συνάρτηση ενεργοποίησης την Ανορθωμένη Γραμμική Μονάδα - ReLU (Rectified Linear Unit). Αμέσως μετά, εφαρμόστηκε η τεχνική «Dropout» η οποία επιτυγχάνει τυχαία ανάθεση, σε κάθε επανάληψη του αλγόριθμου εκπαίδευσης. Το 20% των ενεργοποιήσεων (Activations) του προηγούμενου επιπέδου γίνεται ίσο με το 0. Η τεχνική αυτή χρησιμοποιείται προκειμένου να αποφευχθεί η «υπερπροσαρμογή» (overfitting) του δικτύου στα δεδομένα εκπαίδευσης. Ακολούθησε η προσθήκη ενός πλήρως συνεδεμένου στρώματος με 64 νευρώνες και με συνάρτηση ενεργοποίησης τη 'ReLU'. Αμέσως μετά, εφαρμόστηκε η τεχνική «Dropout» με 30% των ενεργοποιήσεων (Activations) του προηγούμενου επιπέδου να γίνεται ίσο με 0. Τέλος, προστέθηκε ένα πλήρως συνδεδεμένο στρώμα (Dense) το οποίο αποτέλεσε το επίπεδο εξόδου του συστήματος. Ως συνάρτηση ενεργοποίησης επιλέχθηκε η «σιγμοειδής συνάρτηση» (sigmoid) η οποία ακολουθεί κατανομή μεταξύ διαστήματος [0,1]. Η δομή της αρχιτεκτονικής του μοντέλου που κατασκευάστηκε μπορεί να ληφθεί μέσω της μεθόδου «summary». 

Έπειτα, ορίστηκαν οι πληροφορίες αναφορικά με τον «Αλγόριθμο Προσαρμογής» (Fitting Algorithm). Η «Συνάρτηση Σφάλματος» (loss function) που ελαχιστοποιείται είναι η κατηγορηματική διασταυρωμένη εντροπία (Categorical Cross-entropy), ενώ ως βελτιστοποιητής επιλέχθηκε ο «Αλγόριθμος Προσαρμοστικής Εκτίμησης Βαρών» (Adaptive Moment Estimation - Adam). Επιπλέον, δίνεται η δυνατότητα ορισμού μίας λίστας από επιπλέον μετρικές οι οποίες θα υπολογίζονται κατά τη διάρκεια εκπαίδευσης. Στη συγκεκριμένη περίπτωση, ως επιπλέον μετρική επιλέχθηκε η ακρίβεια. Η παραπάνω διαδικασία, πραγματοποιήθηκε μέσω της μεθόδου «compile» του μοντέλου που δημιουργήθηκε. 

```python
# Building Model
densenet201Model = DenseNet201(weights="imagenet", include_top=False,
    input_tensor=Input(shape=(224, 224, 3)))

outputs = densenet201Model.output

outputs = GlobalAveragePooling2D()(outputs)
outputs = Flatten(name="flatten")(outputs)
outputs = Dense(128, activation='relu')(outputs)
outputs = Dropout(0.2)(outputs)
outputs = Dense(64, activation='relu')(outputs)
outputs = Dropout(0.3)(outputs)
outputs = Dense(2, activation="sigmoid")(outputs)

model = Model(inputs=densenet201Model.input, outputs=outputs)

for layer in densenet201Model.layers:
    layer.trainable = False

model.compile(
        loss='categorical_crossentropy', 
        optimizer='adam', 
        metrics=['accuracy']
)
```

## Εμπλουτισμός Δεδομένων / Data Augmentation
Σε προβλήματα με λίγα δεδομένα, όπως στις ιατρικές απεικονίσεις, είναι πολύ εύκολο το μοντέλο να μάθει να ταξινομεί καλύτερα από όσο απαιτείται ζημιώνοντας την ικανότητα γενίκευσής του σε άγνωστα δεδομένα. Το πρόβλημα αυτό καλείται «υπερπροσαρμογή (overfitting)» και μία τεχνική καταπολέμησης του αποτελεί ο εμπλουτισμός του συνόλου εκπαίδευσης (training set) με δεδομένα που έχουν υποστεί ελαφριές παραμορφώσεις, έχοντας ως σκοπό το δίκτυο να μη βλέπει τις ίδιες εικόνες σε πολλές «εποχές» (epochs). Στο Keras αυτό επιτυγχάνεται με τη χρήση της «ImageDataGenerator» η οποία δέχεται μία «ομάδα παρατηρήσεων» (batch), εφαρμόζει τυχαίες γεωμετρικές παραμορφώσεις και επιστρέφει μία νέα ομάδα παρατηρήσεων για να εκπαιδευτεί το δίκτυο. 

```python
# Image Augmentation
# To train on images at different positions, angles, flips, e.t.c.
train_aug = ImageDataGenerator(
    rotation_range=20,
    width_shift_range=0.2,
    height_shift_range=0.2,
    horizontal_flip=True
)
```

## Εκπαίδευση του Νευρωνικού Δικτύου
Με το πέρας των παραπάνω βημάτων, έγινε η έναρξη της εκπαίδευσης του νευρωνικού δικτύου που κατασκευάστηκε προηγουμένως. Η εκπαίδευση του δικτύου, πραγματοποιήθηκε με τη χρήση της μεθόδου «fit» του μοντέλου. Οι παράμετροι «epochs» και «batch size» παίζουν σημαντικό ρόλο στην εκπαίδευση του δικτύου. Ειδικότερα, η παράμετρος «batch size» αποτελεί το μέγεθος της ομάδας παρατηρήσεων και δηλώνει ότι σε κάθε επανάληψη, ο αλγόριθμος βελτιστοποίησης επιλέγει τυχαία μια ομάδα παρατηρήσεων πληθικότητας ίση με την τιμή της παραμέτρου, για τον υπολογισμό της παραγώγου (gradient). Από την άλλη πλευρά, η παράμετρος «epochs» στην ουσία δηλώνει το πλήθος των εποχών, όπου μία εποχή αντιστοιχεί σε ένα πέρασμα όλων των δεδομένων εκπαίδευσης από το νευρωνικό δίκτυο. Στην συγκεκριμένη υλοποίηση ως επιπλέον μετρική έχει καθοριστεί η ακρίβεια (accuracy). 

Συνοπτικά, επιλέχθηκε η εκπαίδευση ενός νευρωνικού δικτύου για 500 εποχές με μέγεθος ομάδας παρατηρήσεων ίσο με 32,  χρησιμοποιώντας ως δεδομένα επαλήθευσης (validation set) τα δεδομένα δοκιμής. Τέλος, το εκπαιδευμένο νευρωνικό δίκτυο αποθηκεύτηκε σε ένα αρχείο με κατάληξη .h5 ώστε να υπάρχει δυνατότητα επαναχρησιμοποίησής του. 

```python
# Τraining config:
epochs = 500
batch_size = 32

# Train the model
history = model.fit(train_aug.flow(X_train, y_train, batch_size=batch_size),
                    validation_data=(X_test, y_test),
                    validation_steps=len(X_test) / batch_size,
                    steps_per_epoch=len(X_train) / batch_size,
                    epochs=epochs)
```

```python
# Save Model and Weights
model.save('densenet201_ct.h5')
model.save_weights('densenet201_weights_ct.hdf5')
```

```python
# Load saved model
model = load_model('densenet201_ct.h5')
```

## Οπτικοποίηση με τη μέθοδο Grad-CAM
Το νευρωνικό δίκτυο ταξινομεί τις εικόνες βάσει ενός διανύσματος χαρακτηριστικών. Στην παρούσα εργασία, χρησιμοποιήθηκε η μέθοδος «Χαρτογράφησης Ενεργοποίησης Κλάσης με Στάθμιση Βαθμίδας» (Grad-guided CAM – Grad CAM). [7] Πιο συγκεκριμένα, η μέθοδος αυτή χρησιμοποιεί τα χαρακτηριστικά που εξάγονται από κάποιο από τα συνελικτικά στρώματα, συνήθως επιλέγεται το τελευταίο συνελικτικό στρώμα, τα οποία χρησιμοποιούνται για την παραγωγή ενός χάρτη ενεργοποίησης. Ο χάρτης ενεργοποίησης αποσκοπεί στον εντοπισμό και στην οπτικοποίηση των σημαντικότερων περιοχών της εικόνας σύμφωνα με το νευρωνικό δίκτυο. Η οπτικοποίηση, λοιπόν, εστιάζει στις περιοχές από τις οποίες εξάγονται αυτά τα χαρακτηριστικά με αποτέλεσμα να παρουσιάζει διττό ρόλο ως προς το πρόβλημα. Συγκεκριμένα, παρέχει σημαντικές πληροφορίες για τη λειτουργία του νευρωνικού, ενώ ταυτόχρονα μπορεί να αποτελέσει ένα ισχυρό μέσο υποβοήθησης της ιατρικής διάγνωσης, καθώς επισημαίνει τις περιοχές στις οποίες εντοπίζεται η ασθένεια. Οι πιο συχνές βλάβες που εντοπίζονται μέσω της αξονικής τομογραφίας θώρακα είναι οι διηθήσεις δίκην θολής υάλου ή/και οζώδεις διηθήσεις στην περιφέρεια των πνευμόνων (Ground-glass opacities - GGO).

```python
# To run the next few cells create a folder with the 'X_train' images before the preprocess
import random

img_paths = ".\\kaggle\\random_images"
img_paths = glob(img_paths + '\*')

img_paths = [(img_paths[index]) for index in random.sample(range(0, len(img_paths)), 5)]
```

```python
val_imgs = []

for i in range(len(img_paths)):
  image = cv2.imread(img_paths[i]) # read file 
  image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB) # arrange format as per kera
  image = cv2.resize(image,(224,224)) # resize as per model
  val_imgs.append(image) # append image
```

```python
# Load the model that was trained above
model = keras.models.load_model("densenet201_ct.h5")

last_conv_layer = "conv5_block32_2_conv"

def get_img_array(img_path, size):
    img = keras.preprocessing.image.load_img(img_path, target_size=size) # `img` is a PIL image of size 224x224
    array = keras.preprocessing.image.img_to_array(img)                  # `array` is a float32 Numpy array of shape (224, 224, 3)
    # We add a dimension to transform our array into a "batch"
    # of size (1, 224, 224, 3)
    array = np.expand_dims(array, axis=0)
    return array

def make_gradcam_heatmap(img_array, model, last_conv_layer_name, pred_index=None):
    # First, create a model that maps the input image to the activations
    # of the last conv layer as well as the output predictions
    grad_model = tf.keras.models.Model(
        [model.inputs], [model.get_layer(last_conv_layer_name).output, model.output]
    )

    # Then, compute the gradient of the top predicted class for our input image
    # with respect to the activations of the last conv layer
    with tf.GradientTape() as tape:
        last_conv_layer_output, preds = grad_model(img_array)
        if pred_index is None:
            pred_index = tf.argmax(preds[0])
        class_channel = preds[:, pred_index]

    # This is the gradient of the output neuron (top predicted)
    grads = tape.gradient(class_channel, last_conv_layer_output)

    # This is a vector where each entry is the mean intensity of the gradient
    # over a specific feature map channel
    pooled_grads = tf.reduce_mean(grads, axis=(0, 1, 2))

    # Multiply each channel in the feature map array
    # by "how important this channel is" with regard to the top predicted class
    # then sum all the channels to obtain the heatmap class activation
    last_conv_layer_output = last_conv_layer_output[0]
    heatmap = last_conv_layer_output @ pooled_grads[..., tf.newaxis]
    heatmap = tf.squeeze(heatmap)

    # For visualization purpose, also normalize the heatmap between 0 & 1
    heatmap = tf.maximum(heatmap, 0) / tf.math.reduce_max(heatmap)
    return heatmap.numpy()
```

```python
# Display
from IPython.display import Image, display
import matplotlib.cm as cm

def save_and_display_gradcam(img_br, heatmap, cam_path="cam.jpg", alpha=0.4):
    heatmap = np.uint8(255 * heatmap)  # Rescale heatmap to a range 0-255

    jet = cm.get_cmap("jet")  # Use jet colormap to colorize heatmap

    jet_colors = jet(np.arange(256))[:, :3]  # Use RGB values of the colormap
    jet_heatmap = jet_colors[heatmap]

    # Create an image with RGB colorized heatmap
    jet_heatmap = keras.preprocessing.image.array_to_img(jet_heatmap)
    jet_heatmap = jet_heatmap.resize((img_br.shape[1], img_br.shape[0]))
    jet_heatmap = keras.preprocessing.image.img_to_array(jet_heatmap)

    # Superimpose the heatmap on original image
    superimposed_img = jet_heatmap * alpha + img
    superimposed_img = np.minimum(superimposed_img, 255.0).astype(np.uint8)  # scale 0 to 255  
    superimposed_img_rgb = keras.preprocessing.image.array_to_img(superimposed_img)
    
    return superimposed_img_rgb
```

```python
# Remove last layer's softmax
model.layers[-1].activation = None
```

```python
# Visualize five random images
plt.figure(figsize=(8, 8))

for i, img in enumerate(val_imgs):
    ax = plt.subplot(1, 5, i+1)
    grad_cam = make_gradcam_heatmap(np.expand_dims(img, axis=0), model, last_conv_layer)
    grad_cam_emphasized = save_and_display_gradcam(img, grad_cam, cam_path="cam.jpg", alpha=0.4)
    plt.axis('off')
    plt.imshow(grad_cam_emphasized)
plt.tight_layout()
```
